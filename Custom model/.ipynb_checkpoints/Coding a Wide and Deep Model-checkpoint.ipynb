```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, concatenate, Input
from tensorflow.keras import Model
from tensorflow.keras.utils import plot_model
```

## Build the Model with functional API

```python
# define inputs
input1 = Input(shape=(1,), name='wide')
input2 = Input(shape=(1,), name='deep')
# define deep path
hidden1 = Dense(32, activation ='relu')(input2)
hidden2 = Dense(16, activation ='relu')(hidden1)
# define merged path
concat = concatenate([input1,hidden2])

# define outputs
output = Dense(1, name='output')(concat)
output2 = Dense(1, name='output2')(hidden2)

model = Model(inputs=[input1,input2], outputs=[output,output2])
```

```python
plot_model(model)
```
